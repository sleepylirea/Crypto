```python
def compter(text):
    # text = text.lower()
    d = {}
    for i in range (len(text)):
        if text[i] not in d :
            d[text[i]]=0
        d[text[i]]+=1
    return d

def compter_occurences(c, text) :
    # text = text.lower()
    x = 0
    for i in range (len(text)):
        if c == text[i]:
            x += 1
    return c, x

def substituer(a, Na, t) :
    t = t.lower()
    L = list(t)
    for i in range (len(L)) :
        if L[i] == a :
            L[i] = Na
    T = ''.join(L)
    return T

def decaler(c, d):
    c = chr(ord(c)+d)
    return c

def chiffrement_cesar(t, d):
    L = list(t)
    for i in range (len(L)) :
        L[i] = decaler(L[i], d)
    T = ''.join(L)
    return (T,d)

def dechiffrement_cesar(t, d):
    L = list(t)
    for i in range (len(L)) :
        L[i] = decaler(L[i], (-d))
    T = ''.join(L)
    return T

def couper(p, n, t) :    # p = parti   n = numÃ©ro dans la partie
    L = list(t)
    if len(L)/p != (len(L)//p) :
        for i in range (len(L)%p) :
            L.append('')
    T = []
    for i in range (len(L)) :
        if i%p == n :
            T.append(L[(i)])
    T = ''.join(T)
    return T

def assembler(L):
    T = []
    n = len(L)
    for i in range (len(L[n-1])) :
        for j in range (n) :
            T.append(L[j][i])
    for loop in range (n) :
        if len(L[0]) == len(L[loop]) :
            T.append(L[loop][len(L[n-1])])
    T = ''.join(T)
    return T

def complet (t, p):        # text   parti   
    T = []
    for i in range (p) :
        T.append(couper(p,i,t))
    for i in range (p) :
        D = compter(T[i])
        a = 0
        b = ''
        for k,v in D.items() :
            if v > a :
                a = v
                b = k
        d = ord(b)-ord(' ')
        T[i] = dechiffrement_cesar(T[i], d)
    return assembler(T)
```
